```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

# 파일 경로 설정
file_path = '/content/drive/MyDrive/데이터마이닝/association_rules.csv'

# CSV 파일 불러오기
df = pd.read_csv(file_path)
```

```python
def all_count_left_exercises(data):
    exercise_counts = {}

    for row in data:
        exercises = row.split("=>")[0].strip().split(",")
        left_count = len(exercises)
        if left_count not in exercise_counts:
            exercise_counts[left_count] = 1
        else:
            exercise_counts[left_count] += 1

    return exercise_counts


result = all_count_left_exercises(df['rules'])

for count, frequency in result.items():
    print(f"왼쪽 운동 {count}개인 것은 총 {frequency}개 있습니다.")
```

```python
from collections import Counter

def count_left_exercises(data):
    left_exercises_count = []

    for row in data:
        row = row.replace("{", "").replace("}", "")  # "{"와 "}" 문자열 제거
        left_exercises, right_exercises = row.split("=>")
        left_exercises = left_exercises.strip().split(",")
        right_exercise = right_exercises.strip()

        if len(left_exercises) == 2:  # 왼쪽 운동이 2개인 경우
            combination = tuple(sorted(left_exercises + [right_exercise]))
            left_exercises_count.append(combination)

    return Counter(left_exercises_count)


result = count_left_exercises(df['rules'])
total_lines = len(result)


for combination, frequency in result.items():
    print(f"{', '.join(combination)} (빈도: {frequency})")
    
print("총",total_lines,"줄이 출력되었습니다.")
```

```python
def count_combinations_with_frequency_above_2(data):
    result = count_left_exercises(data)
    count = 0

    for frequency in result.values():
        if frequency == 2:
            count += 1

    return count


frequency_above_2_count = count_combinations_with_frequency_above_2(df['rules'])
print("빈도가 2 인 조합의 개수:", frequency_above_2_count)
```

```python
def count_combinations_with_frequency_above_3(data):
    result = count_left_exercises(data)
    count = 0

    for frequency in result.values():
        if frequency >= 3:
            count += 1

    return count


frequency_above_3_count = count_combinations_with_frequency_above_3(df['rules'])
print("빈도가 3 인 조합의 개수:", frequency_above_3_count)
```

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

# 파일 경로 설정
file_path = ('/content/drive/MyDrive/데이터마이닝/데이터셋/main_exercises.csv')

# CSV 파일 불러오기
df = pd.read_csv(file_path)

# exercises가 [''] 값을 가지는 행 제거
df = df[df['exercises'] != "['']"]

#각 행에 대한 결측치 비율
empty_ratio = df.isnull().mean() * 100

#해당 열의 결측치 비율이 99프로가 넘어가는거 싹다 지우기
columns_to_drop = empty_ratio[empty_ratio > 1].index
df = df.drop(columns=columns_to_drop)

#결측 열을 날린 데이터셋에 대하여 각 행에 남아있는 결측치 지우기
for column in df.columns:
    if df[column].isnull().any():
        df = df.dropna(subset=[column])
```

```python
#살아남은 어트리뷰트
for column_name in df.columns:
    print(column_name)
```

```python
#각 행에 대한 결측치 비율 다시한번 확인
empty_ratio = df.isnull().mean() * 100
print(empty_ratio)
```

```python
import ast
# 'exercises' 열의 문자열을 리스트 형태로 변환
df['exercises'] = df['exercises'].apply(ast.literal_eval)
```

```python
# 'exercises' 열의 리스트를 집합 형식으로 변환
df['exercises'] = df['exercises'].apply(lambda x: set(x))
```

```python
df['exercises']
```

```python
#set으로 exercises와 최종_ard의 0번째행부터 28번째 행까지 
# 쭉 확인하면서 새로운 df['라벨링']에 리스트 형태로 운동을 저장시키는 코드
df['라벨링'] = df['exercises'].apply(lambda x: [i for i, ard in enumerate(df_ard['최종_ard']) if set(ard).issubset(x)])
```

```python
# 나중에 분석할때 편하라고
# 최종_ard를 더미 변수화하여 라벨링 뒤에다가 라벨링_더미0, 라벨링_더미1, ... 컬럼 생성해놓은거
dummy_columns = []
for i, ard in enumerate(df_ard['최종_ard']):
    dummy_column_name = f'라벨링_더미{i}'
    df[dummy_column_name] = df['exercises'].apply(lambda x: 1 if set(ard).issubset(x) else 0)
    dummy_columns.append(dummy_column_name)
```

```python
#일단 더미변수 빼놓고 라벨링만 출력해서 확인
print(df[['exercises', '라벨링']])
```

```python
#라벨링 중 해당되는게 없어 [] 이런식으로 출력된거 없애는 코드
df_filtered = df[df['라벨링'].apply(lambda x: len(x) > 0)]

# 결과 출력
print(df_filtered[['exercises', '라벨링']])
```
