```python
import pandas as pd
```

```python
df1 = pd.read_csv('EXERCISE_PRESCRIPTION_DATA_1.csv')
```

```python
df2 = pd.read_csv('EXERCISE_PRESCRIPTION_DATA_2.csv')
```

```python
df = pd.concat([df1,df2])
```

```python
df.describe(include='all')
```

```python
program = df['MVM_PRSCRPTN_CN']

```

```python
all_list = []
for string in program:
    string = str(string)
    string = string.replace(' ',"")
    split_list = string.split('/')
    all_list.append(split_list)
    
```

```python
len(all_list)
```

```python
new_list = []
for each_list in all_list:
    tempCount = 0
    for i in range(len(each_list)):
        each_exer = each_list[i]        
        if (each_exer[0]=='본'):
            new_list.append(each_exer)
            tempCount += 1
        if (i == (len(each_list)-1)) and (tempCount==0) :
            new_list.append('Null')
            
        
            
            
len(new_list)
            
    
    
```

```python
main_program_split = []

for string in new_list:
    string = string.replace('본운동:',"")
    string = string.replace('Null',"")
    temp_list = string.split(',')
    main_program_split.append(temp_list)
    
len(main_program_split)
```

```python
df['exercises'] = main_program_split
```

```python
df['exercises']
```

```python
counts = {}
for each in main_program_split:
    for x in each:
        if x in counts:
            counts[x] += 1
        else:
            counts[x] = 1

counts
```

## 내림차순으로 정리한 각 운동이 transaction 데이터에서 얼마나 등장했는지

```python
counts_sorted = sorted(counts.items(), key=lambda x: x[1], reverse=True)
counts_sorted
```
